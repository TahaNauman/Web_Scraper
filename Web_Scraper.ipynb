```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import json

# Fetch Page
def fetch_page(url):
    try:
        ua = UserAgent().random
    except:
        ua = "Mozilla/5.0"

    headers = {
        "User-Agent": ua
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            return response.text
        else:
            print(f" Failed to fetch page. Status code: {response.status_code}")
            return None
    except Exception as e:
        print("Error:", e)
        return None


# Get all unique tags
def get_all_tags(html):
    soup = BeautifulSoup(html, "html.parser")
    tags = set()

    for tag in soup.find_all():
        tags.add(tag.name)

    return sorted(tags)


#Extract selected tags
def extract_tags(html, selected_tags):
    soup = BeautifulSoup(html, "html.parser")
    results = {}

    for tag in selected_tags:
        elements = soup.find_all(tag)

        extracted = []
        for el in elements:
            text = el.get_text(separator="",strip=True)

            if text:  # avoid emptry entries
                extracted.append(text)

        results[tag] = extracted

    return results


#Save to file
def save_data(data, filename="output.json"):
  try:
      with open(filename, "w", encoding="utf-8") as f:
          json.dump(data, f, indent=4, ensure_ascii=False)
      print(f" Data saved to {filename}")
  except Exception as e:
      print("Error saving file:", e)

#Main Program
def main():
    url = input("Enter URL: ").strip()

    html = fetch_page(url)
    if not html:
        return

    # Show available tags
    tags = get_all_tags(html)

    print("\n Available Tags:")
    print(", ".join(tags))

    # User selects tags
    selected = input("\nEnter tags to extract (comma separated): ").split(",")
    selected = [t.strip() for t in selected if t.strip()]

    if not selected:
        print(" No valid tags entered")
        return

    # Extract data
    data = extract_tags(html, selected)

    # Save output
    save_data(data)


# Run script
if __name__ == "__main__":
    main()
```
